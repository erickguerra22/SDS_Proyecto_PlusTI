### Ingeniería de Variables

```python
df_featured = pd.read_csv("./feature_engineering_work/dataset_feature_engineering.csv")
```

```python
# 1. Características basadas en tiempo
df_featured['transaction_dt'] = pd.to_datetime(df_featured['unix_time'], unit='s')
df_featured['hour_of_day'] = df_featured['transaction_dt'].dt.hour
df_featured['day_of_week'] = df_featured['transaction_dt'].dt.dayofweek
df_featured['is_weekend'] = df_featured['day_of_week'].isin([5, 6]).astype(int)
df_featured['is_night'] = ((df_featured['hour_of_day'] >= 22) | (df_featured['hour_of_day'] <= 5)).astype(int)
df_featured['day_of_month'] = df_featured['transaction_dt'].dt.day
df_featured['week_of_month'] = (df_featured['day_of_month'] - 1) // 7 + 1
df_featured['time_of_day'] = pd.cut(
    df_featured['hour_of_day'],
    bins=[-1, 5, 11, 16, 21, 24],
    labels=['late_night', 'morning', 'afternoon', 'evening', 'early_night']
)
```

```python
# 2. Edad del titular de la tarjeta
df_featured['dob_dt'] = pd.to_datetime(df_featured['dob'])
df_featured['age'] = df_featured['year'] - df_featured['dob_dt'].dt.year
```

```python
# 3. Características basadas en monto
amt_percentiles = df_featured['amt'].quantile([0.25, 0.5, 0.75, 0.9, 0.95, 0.99])
df_featured['amt_category'] = pd.cut(
    df_featured['amt'],
    bins=[0, amt_percentiles[0.25], amt_percentiles[0.5], amt_percentiles[0.75], 
          amt_percentiles[0.9], amt_percentiles[0.95], amt_percentiles[0.99], float('inf')],
    labels=['muy_bajo', 'bajo', 'medio', 'alto', 'muy_alto', 'extremo', 'outlier']
)
```

```python
df_featured['amt_to_monthly_ratio'] = df_featured['amt'] / df_featured['amt_month'].replace(0, 1)
df_featured['amt_to_yearly_ratio'] = df_featured['amt'] / df_featured['amt_year'].replace(0, 1)

category_avg_amt = df_featured.groupby('category')['amt'].transform('mean')
df_featured['amt_deviation_from_category_avg'] = df_featured['amt'] - category_avg_amt

df_featured['amt_percentile_in_category'] = df_featured.groupby('category')['amt'].transform(
    lambda x: x.rank(pct=True)
)
```

```python
# 4. Características basadas en distancia
df_featured['distance_bin'] = pd.cut(
    df_featured['dist_between_client_and_merch'],
    bins=[0, 10, 50, 100, 500, float('inf')],
    labels=['muy_cerca', 'cerca', 'medio', 'lejos', 'muy_lejos']
)

df_featured['distance_amount_ratio'] = df_featured['dist_between_client_and_merch'] / (df_featured['amt'] + 1)

merchant_avg_dist = df_featured.groupby('merchant')['dist_between_client_and_merch'].transform('mean')
merchant_std_dist = df_featured.groupby('merchant')['dist_between_client_and_merch'].transform('std').fillna(0)
df_featured['dist_zscore'] = (df_featured['dist_between_client_and_merch'] - merchant_avg_dist) / (merchant_std_dist + 1)
df_featured['is_unusual_distance'] = (df_featured['dist_zscore'].abs() > 2).astype(int)
```

```python
# 5. Características basadas en frecuencia de transacciones
df_featured['days_since_last_purchase'] = 30 / (df_featured['times_shopped_at_merchant_month'] + 1)

df_featured['trans_freq_ratio'] = df_featured['times_shopped_at_merchant'] / (df_featured['times_shopped_at_merchant_year'] + 1)
df_featured['trans_freq_month_ratio'] = df_featured['times_shopped_at_merchant_month'] / (df_featured['times_shopped_at_merchant'] + 1)
df_featured['trans_freq_day_ratio'] = df_featured['times_shopped_at_merchant_day'] / (df_featured['times_shopped_at_merchant_month'] + 1)
```

```python
# 6. Características basadas en comerciantes
df_featured['merchant_type'] = df_featured['merchant'].apply(lambda x: x.split('_')[0] if '_' in str(x) else x)
merchant_cat_counts = df_featured.groupby('merchant')['category'].transform('nunique')
df_featured['merchant_multi_category'] = (merchant_cat_counts > 1).astype(int)
```

```python
# 7. Características demográficas
df_featured['gender_encoded'] = df_featured['gender'].map({'M': 0, 'F': 1})
```

```python
df_featured['pop_category'] = pd.cut(
    df_featured['city_pop'],
    bins=[0, 1000, 10000, 100000, float('inf')],
    labels=['rural', 'pequeña_ciudad', 'suburbano', 'urbano']
)
```

```python
# 8. Interacciones y características compuestas
df_featured['first_time_distance_interaction'] = df_featured['first_time_at_merchant'].astype(int) * df_featured['dist_between_client_and_merch']
```

```python
df_featured['first_time_amount_interaction'] = df_featured['first_time_at_merchant'].astype(int) * df_featured['amt']

df_featured['transaction_uniqueness_score'] = (
    (df_featured['first_time_at_merchant'].astype(int) * 0.5) +
    (df_featured['dist_between_client_and_merch'] > 100).astype(int) * 0.3 +
    (df_featured['amt'] > df_featured['amt_month'] * 0.5).astype(int) * 0.2
)
```

```python
# 9. Características basadas en actividad de compras netas
df_featured['shopping_net_ratio'] = df_featured['amt_month_shopping_net_spend'] / (df_featured['amt_month'] + 0.01)
```

```python
# 10. Características geoespaciales
df_featured['lat_diff'] = (df_featured['lat'] - df_featured['merch_lat']).abs()
df_featured['long_diff'] = (df_featured['long'] - df_featured['merch_long']).abs()
```

```python
# 11. Características estadísticas de la actividad
client_avg_amt = df_featured.groupby('cc_num')['amt'].transform('mean')
client_std_amt = df_featured.groupby('cc_num')['amt'].transform('std').fillna(0)
df_featured['amt_zscore'] = (df_featured['amt'] - client_avg_amt) / (client_std_amt + 1)
df_featured['is_unusual_amount'] = (df_featured['amt_zscore'] > 2).astype(int)
```

```python
# 12. Velocidad de transacciones
client_trans_per_day = df_featured.groupby(['cc_num', 'trans_day', 'trans_month', 'year']).size().reset_index(name='client_daily_trans_count')
df_featured = pd.merge(df_featured, client_trans_per_day, on=['cc_num', 'trans_day', 'trans_month', 'year'], how='left')
```

```python
# 13. Variables dummy para categorías importantes
category_dummies = pd.get_dummies(df_featured['category'], prefix='cat')
df_featured = pd.concat([df_featured, category_dummies], axis=1)
```

```python
time_dummies = pd.get_dummies(df_featured['time_of_day'], prefix='time')
df_featured = pd.concat([df_featured, time_dummies], axis=1)

distance_dummies = pd.get_dummies(df_featured['distance_bin'], prefix='dist')
df_featured = pd.concat([df_featured, distance_dummies], axis=1)

pop_dummies = pd.get_dummies(df_featured['pop_category'], prefix='pop')
df_featured = pd.concat([df_featured, pop_dummies], axis=1)
```

```python
# 14. Patrones temporales semanales
df_featured['day_name'] = df_featured['day_of_week'].map({
    0: 'Lunes', 1: 'Martes', 2: 'Miércoles', 3: 'Jueves', 
    4: 'Viernes', 5: 'Sábado', 6: 'Domingo'
})
```

```python
# 15. Ratio de transacciones de primera vez
df_featured['first_time_merchant_ratio'] = df_featured.groupby('cc_num')['first_time_at_merchant'].transform('mean')
```

```python
# 16. Características basadas en el trabajo del cliente
def categorize_job(job):
    if pd.isna(job):
        return 'unknown'
    job = job.lower()
    if any(term in job for term in ['developer', 'engineer', 'programmer', 'analyst', 'scientist', 'it']):
        return 'tech'
    elif any(term in job for term in ['doctor', 'nurse', 'therapist', 'health', 'medical', 'psychologist']):
        return 'healthcare'
    elif any(term in job for term in ['teacher', 'professor', 'education', 'academic']):
        return 'education'
    elif any(term in job for term in ['manager', 'director', 'executive', 'ceo', 'cfo', 'officer']):
        return 'management'
    elif any(term in job for term in ['lawyer', 'attorney', 'legal', 'advocate']):
        return 'legal'
    elif any(term in job for term in ['designer', 'artist', 'creative', 'writer']):
        return 'creative'
    elif any(term in job for term in ['consultant', 'advisor', 'specialist']):
        return 'consultant'
    else:
        return 'other'

df_featured['job_category'] = df_featured['job'].apply(categorize_job)
job_dummies = pd.get_dummies(df_featured['job_category'], prefix='job')
df_featured = pd.concat([df_featured, job_dummies], axis=1)
```

```python
# 17. Score compuesto de riesgo de fraude
df_featured['fraud_risk_score'] = (
    (df_featured['first_time_at_merchant'].astype(int) * 0.3) +
    (df_featured['dist_between_client_and_merch'] > 100).astype(int) * 0.2 +
    (df_featured['amt'] > df_featured['amt_month'] * 0.7).astype(int) * 0.15 +
    (df_featured['is_night'] * 0.1) +
    (df_featured['is_weekend'] * 0.05) +
    (df_featured['is_unusual_amount'] * 0.2)
)
```

```python
# 18. Entrópica de patrones de transacción
def calculate_entropy(group):
    if len(group) <= 1:
        return 0
    amounts = group['amt'].values
    
    probs = amounts / amounts.sum()
    entropy = -np.sum(probs * np.log2(probs + 1e-10))
    return entropy

client_entropy = df_featured.groupby('cc_num').apply(calculate_entropy).reset_index(name='transaction_entropy')
df_featured = pd.merge(df_featured, client_entropy, on='cc_num', how='left')
```
